```python
# Import dependencies
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
# Load Data
finance_1617_load = "../Resources/Finance Data/School Budget Data '16-'17.csv"
finance_1516_load = "../Resources/Finance Data/School Budget Data '15-'16.csv"
finance_1415_load = "../Resources/Finance Data/School Budget Data '14-'15.csv"
finance_1314_load = "../Resources/Finance Data/School Budget Data '13-'14.csv"
finance_1213_load = "../Resources/Finance Data/School Budget Data '12-'13.csv"
```

```python
# Store 16-17 Data into DataFrame
finance_1617_df = pd.read_csv(finance_1617_load, 
                              names=['Ccode', 'Dcode', 'SchoolCode', 'Fiscalyear', 'Period','Colcode','Account','Fund',
                                     'Resource','Projectyear','Goal','Function','Object','Value'])
# Delete values with Function value 0
finance_1617_df = finance_1617_df[finance_1617_df.Function != 0]
finance_1617_df

finance_1617_df
```

```python
# Store 15-16 Data into DataFrame
finance_1516_df = pd.read_csv(finance_1516_load,
                              names=['Ccode', 'Dcode', 'SchoolCode', 'Fiscalyear', 'Period','Colcode','Account','Fund',
                                     'Resource','Projectyear','Goal','Function','Object','Value'])
# Delete values with Function value 0
finance_1516_df = finance_1516_df[finance_1516_df.Function != 0]
finance_1516_df

finance_1516_df
```

```python
# Store 14-15 Data into DataFrame
finance_1415_df = pd.read_csv(finance_1415_load,
                              names=['Ccode', 'Dcode', 'SchoolCode', 'Fiscalyear', 'Period','Colcode','Account','Fund',
                                     'Resource','Projectyear','Goal','Function','Object','Value'])
# Delete values with Function value 0
finance_1415_df = finance_1415_df[finance_1415_df.Function != 0]
finance_1415_df

finance_1415_df
```

```python
# Store 13-14 Data into DataFrame
finance_1314_df = pd.read_csv(finance_1314_load,
                              names=['Ccode', 'Dcode', 'SchoolCode', 'Fiscalyear', 'Period','Colcode','Account','Fund',
                                     'Resource','Projectyear','Goal','Function','Object','Value'])
# Delete values with Function value 0
finance_1314_df = finance_1314_df[finance_1314_df.Function != 0]
finance_1314_df

finance_1314_df
```

```python
# Store 12-13 Data into DataFrame
finance_1213_df = pd.read_csv(finance_1213_load,
                               names=['Ccode', 'Dcode', 'SchoolCode', 'Fiscalyear', 'Period','Colcode','Account','Fund',
                                     'Resource','Projectyear','Goal','Function','Object','Value'])
# Delete values with Function value 0
finance_1213_df = finance_1213_df[finance_1213_df.Function != 0]
finance_1213_df

finance_1213_df
```

```python
# Set the x-axis to a list of the years under analysis
x_axis = ["2012-2013", "2013-2014", "2015-2015", "2015-2016", "2016-2017"]

# Find the sum for each year
sum_1213 = finance_1213_df["Value"].sum() / 1000000000
sum_1314 = finance_1314_df["Value"].sum() / 1000000000
sum_1415 = finance_1415_df["Value"].sum() / 1000000000
sum_1516 = finance_1516_df["Value"].sum() / 1000000000
sum_1617 = finance_1617_df["Value"].sum() / 1000000000

# Set the y-axis to the sum of values for each year
y_axis = [sum_1213.round(2), sum_1314.round(2), sum_1415.round(2), sum_1516.round(2), sum_1617.round(2)]

y_axis
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="green", label='Total Spend')

# Create labels for the x an y axes
plt.xlabel("Academic School Years")
plt.ylabel("Total Spend (Billions)")

# Create a title
plt.title("California Spend by School Year")

# Add the legend
plt.legend()
```
