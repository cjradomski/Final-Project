```python
#Add the Pandas dependency
import pandas as pd
import zipfile
```

```python
#Files to load
zf=zipfile.ZipFile("../enrollment '14-'15.zip")
enroll_df=pd.read_csv(zf.open("enrollment '14-'15.csv"))
enroll_df
```

```python
enroll_df=enroll_df.sort_values(by='CDS_CODE')
enroll_df.head()
```

```python
#Determine any missing values 
enroll_df.dropna()
enroll_df.count()
```

```python
#Determine data types
enroll_df.dtypes
```

```python
clean_enroll_df=enroll_df[['CDS_CODE','DISTRICT','ETHNIC','GENDER','GR_12','ENR_TOTAL']]
clean_enroll_df=clean_enroll_df.reset_index(drop=True)
clean_enroll_df
```

```python
clean_enroll_df['CDS_CODE'] = clean_enroll_df['CDS_CODE'].astype('int64') 
clean_enroll_df
```

```python
clean_enroll_df=clean_enroll_df.groupby(['CDS_CODE','ETHNIC','GENDER']).sum()
clean_enroll_df
```

```python
clean_enroll_df.to_csv('../Enroll_Data/clean_enroll14_15.csv',index=True)
```
