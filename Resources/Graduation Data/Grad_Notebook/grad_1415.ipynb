```python
#add dependencies
import pandas as pd
```

```python
#Load file
grad_df=pd.read_csv("../Graduation '14-'15.zip")
grad_df.head()
```

```python
#Determine any missing values
grad_df.dropna()
grad_df.count()
```

```python
#Determine data types
grad_df.dtypes
```

```python
#keep wanted columns
grad_df=grad_df[['CDS_CODE','ETHNIC','GRADS','UC_GRADS']]
grad_df
```

```python
#convert CDS_CODE to int
grad_df['CDS_CODE'] = grad_df['CDS_CODE'].astype('int64') 
grad_df.head(11)
```

```python
#Group and sort by CDS and ethnic
grad_df=grad_df.groupby(['CDS_CODE','ETHNIC']).sum()
grad_df=grad_df.sort_values(by=['CDS_CODE','ETHNIC'])
grad_df.head(11)
```

```python
grad_df.to_csv('../Grad_Data/grad_1415.csv')
```
